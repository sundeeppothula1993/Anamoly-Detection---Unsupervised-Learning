```python
df.head()
```

```python
doctorss = df.providerid.unique()
cost2 = {}
for e in doctorss:
    sliced_df = df.loc[df['providerid'] == e]
    cost2[e] = sliced_df.diff_percent.mean()

df['final_avg'] = df.providerid.apply(lambda x: cost2[x])
```
